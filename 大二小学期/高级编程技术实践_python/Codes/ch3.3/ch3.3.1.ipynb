```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [5, 4, 2, 1]  # x轴坐标数据
y = [7, 8, 9, 10]  # y轴坐标数据
# 绘制线段
plt.plot(x, y)
```

```python
# 设置中文字体
plt.rcParams['font.family'] = ['SimHei']
# 绘制线段
plt.plot(x, y)
plt.title('绘制折线图')  # 添加图表标题
```

```python
# 绘制线段
plt.plot(x, y)
plt.ylabel('y轴')  # 添加y轴标题
plt.xlabel('x轴')  # 添加x轴标题
```

```python
# 绘制线段
plt.plot(x, y, label='线1')
plt.ylabel('y轴')  # 添加y轴标题
plt.xlabel('x轴')  # 添加x轴标题
plt.legend()  # 设置图例
```

```python
# 设置图表大小
plt.figure(figsize=(10, 5))

# 绘制线段
plt.plot(x, y, label='线1')
plt.ylabel('y轴')  # 添加y轴标题
plt.xlabel('x轴')  # 添加x轴标题
plt.legend()  # 设置图例
```
