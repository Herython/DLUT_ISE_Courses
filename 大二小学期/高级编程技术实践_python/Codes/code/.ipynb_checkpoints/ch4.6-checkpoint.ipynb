```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
L=[3, 2, 0, 1, 4]
sns.violinplot(y=L)
```

```python
s = pd.Series(L)

print('中位数', s.median())
print('上四分位数', s.quantile(0.25))
print('中位数', s.quantile(0.5))
print('下四分位数', s.quantile(0.75))
```

```python
tips = sns.load_dataset('tips')
sns.boxplot(x='day', y='tip', hue='sex', data=tips)
```
