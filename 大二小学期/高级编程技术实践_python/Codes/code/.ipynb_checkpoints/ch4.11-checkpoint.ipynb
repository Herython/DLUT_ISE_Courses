```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
# 设置中文
sns.set_style("darkgrid", {'font.family':'simhei'})
```

```python
df = sns.load_dataset('tips')
sns.regplot(x='total_bill', y='tip', data=df)
```

```python
file_path='data\\'
df20 = pd.read_excel(file_path+'全国总人口数据.xls', sheet_name='20年数据',skiprows=2,skipfooter=2)
df20["时间"] = df20["时间"].map(lambda x: int(x[0:-1]))
```

```python
plt.figure(figsize=(10,10))
sns.regplot(x = "时间", y = "年末总人口(万人)", data = df20, color="g")
```
