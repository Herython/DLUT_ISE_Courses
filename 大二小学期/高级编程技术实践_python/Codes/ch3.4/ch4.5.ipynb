```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

### 4.5.1 	关联图——散点图

```python
n = 1024
x = np.random.normal(0, 1, n)
y = np.random.normal(0, 1, n)
sns.scatterplot(x=x, y=y)
plt.title('绘制散点图', fontproperties='SimHei')
```

```python
sns.set_style("darkgrid", {'font.family':'simhei'})
file_path='data\\'
df20 = pd.read_excel(file_path+'全国总人口数据.xls', sheet_name='20年数据',skiprows=2,skipfooter=2)
df20 = df20.sort_values('时间') 
plt.figure(figsize=(15, 6))
sns.scatterplot(x='时间', y='年末总人口(万人)', data=df20)
```

```python
tips = sns.load_dataset('tips')
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=tips)
```

### 4.5.3 	分面网格关联图

```python
# 加载seaborn数据集mpg（小汽车每加仑在城市公路行驶英里数）
# 字段说明
#     1. mpg:          每加行驶仑英里数
#     2. cylinders:    汽缸数量
#     3. displacement: 排量
#     4. horsepower:   马力 
#     5. weight:       重量 
#     6. acceleration: 加速度
#     7. model year:   生产年份
#     8. origin:       原产地
#     9. car name:     车名
mpg_df = sns.load_dataset("mpg")
```

```python
g = sns.relplot(x="displacement", y="mpg", data=mpg_df)
```

```python
sns.catplot(x='day', y='tip', hue='sex',col='time', data=tips)
```

```python
g = sns.relplot(x="displacement", y="mpg", col='cylinders', data=mpg_df)
```

```python
g = sns.relplot(x="displacement", y="mpg", col='cylinders', row='origin', data=mpg_df)
```

```python
g = sns.relplot(x="displacement", y="mpg", col='cylinders', hue='origin', data=mpg_df)
```

```python
g = sns.relplot(x="displacement", y="mpg", col='cylinders', col_wrap=3, hue='origin', data=mpg_df)
```

```python
g = sns.relplot(x="displacement", y="mpg", col='cylinders', col_wrap=3, hue='origin',kind="line",  data=mpg_df)
```
