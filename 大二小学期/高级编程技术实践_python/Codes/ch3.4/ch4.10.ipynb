```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
# 设置样式
sns.set_style('darkgrid')
```

```python
mpg_df = sns.load_dataset('mpg')
```

```python
# 1、使用Seaborn库内置数据集mpg绘制关联散点图。
sns.scatterplot(data=mpg_df, x='mpg', y='weight', hue='origin')
```

```python
# 2、使用Seaborn库内置数据集mpg绘制关联线图。
sns.lineplot(data=mpg_df, x='mpg', y='weight', hue='origin')
```

```python
# 3、使用Seaborn库内置数据集mpg绘制Dist分布图。
sns.distplot(mpg_df['mpg'])
```

```python
# 4、使用Seaborn库内置数据集mpg绘制KDE密度图。
sns.kdeplot(mpg_df['mpg'], mpg_df['displacement'])
```

```python
# 5、使用Seaborn库内置数据集mpg绘制联合图。
sns.jointplot(x='displacement', y='mpg', data=mpg_df)
```

```python
# 6、使用Seaborn库内置数据集mpg绘制箱线图。
sns.boxplot(x='cylinders', y='mpg', hue='origin', data=mpg_df) 
```

```python
# 7、使用Seaborn库内置数据集mpg绘制小提琴图。
sns.violinplot(x='cylinders', y='mpg', hue='origin', data=mpg_df) 
```

```python
# 8、使用Seaborn库内置数据集mpg绘制柱状图。
sns.barplot(x='cylinders', y='mpg', hue='origin', data=mpg_df) 
```

```python
# 9、使用Seaborn库内置数据集mpg绘制Strip散点图。
sns.stripplot(x="cylinders", y="mpg", hue="origin", data=mpg_df) 
```

```python
# 10、使用Seaborn库内置数据集mpg绘制Swarm散点图。
sns.swarmplot(x='cylinders', y='mpg', hue='origin', data=mpg_df) 
```

```python
# 11、使用Seaborn库内置数据集mpg绘制线性回归图。
sns.regplot(x='displacement', y='mpg', data=mpg_df)
```

```python
# 12、使用Seaborn库内置数据集mpg绘制分面网格线性回归图。
sns.lmplot(x='displacement', y='mpg', col='origin', data=mpg_df)
```

```python
# 13、使用Seaborn库内置数据集mpg绘制分面网格分类图。
sns.catplot(x='cylinders', y='mpg', col='origin', data=mpg_df, kind='swarm')
```

```python
tips = sns.load_dataset('tips')
```

```python
# 14、使用Seaborn库内置数据集tips绘制,通过FacetGrid类绘制分面网格密度图。
g = sns.FacetGrid(tips, col='time')
g.map(sns.distplot, 'tip')
```

```python
# 15、使用Seaborn库内置数据集tips绘制,通过FacetGrid类绘制分面网格线图。
g = sns.FacetGrid(tips, col='time', row='smoker')
g.map(sns.lineplot, 'total_bill', 'tip')
```

```python
# 16、使用Seaborn库内置数据集tips绘制,通过FacetGrid类绘制分面网格线性回归图。
g = sns.FacetGrid(tips, col='time', row='smoker')
g.map(sns.regplot, 'total_bill', 'tip')
```
